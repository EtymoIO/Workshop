# Text Extraction

```python
path_to_pdf = 'ipython_files/data/example1.pdf'
```

## pdftotxt

```python
import os

cmd = "pdftotext %s %s" % (path_to_pdf, 'ipython_files/example.txt')
os.system(cmd)
with open('ipython_files/example.txt', 'r', encoding="latin-1") as myfile:
    text_pdftotxt = myfile.read().replace('\n', '')
    
print(text_pdftotxt)
```

## Textract (Backend pdftotxt)

```python
import textract

text_textract_pdftotxt = textract.process(path_to_pdf)
text_textract_pdftotxt = text_textract_pdftotxt.decode(encoding="latin-1")

print(text_textract_pdftotxt)
```

## PyPDF2

```python
from PyPDF2 import PdfFileReader

pdf = PdfFileReader(path_to_pdf, strict = False)
text_pypdf2 = ''

if pdf.isEncrypted:
    pdf.decrypt('')
    
page_num = 0
for page in pdf.pages:
    page_num += 1
    data = page.extractText()
    text_pypdf2 += data
    
print(text_pypdf2)
```

## pdfminer

```python
from pdfminer.converter import TextConverter
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.layout import LAParams
from io import StringIO


text_pdfminer = ''

laparams = LAParams()
laparams.all_texts = True  
rsrcmgr = PDFResourceManager()

page_num = 0
fp = open(path_to_pdf,'rb')
for page in PDFPage.get_pages(fp):
    page_num += 1
    retstr = StringIO()
    device = TextConverter(rsrcmgr, retstr, codec='utf-8', laparams=laparams)
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    interpreter.process_page(page)
    data = retstr.getvalue()
    retstr.close()
    
    text_pdfminer += data
    
print(text_pdfminer)
```

## pyocr (backend tesseract)

```python
from PIL import Image as Img
from wand.image import Image
import pyocr
import pyocr.builders
import sys
import glob
import os
import os.path as p
import shutil
import time

tools = pyocr.get_available_tools()
if len(tools) == 0:
    print("No OCR tool found")
    sys.exit(1)

tool = tools[0]
lang = tool.get_available_languages()[0]

with Image(filename=path_to_pdf, resolution=300) as img:
    #keep good quality
    img.compression_quality = 80
    #save it to tmp name
    img.save(filename='ipython_files/img/example.jpg')   

txt = ''
for images in sorted(os.listdir('ipython_files/img')):
    txt += tool.image_to_string(
        Img.open('ipython_files/img/' + images),
        lang=lang,
        builder=pyocr.builders.TextBuilder())

shutil.rmtree('ipython_files/img')
os.mkdir('ipython_files/img')
text_tesseract = txt

print(text_tesseract)
```
